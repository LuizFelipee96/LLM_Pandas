```python
import pandas as pd
from pandasai import PandasAI
from pandasai.llm.openai import OpenAI
```

```python
#Criar o dataframe

amputee_df = pd.DataFrame({
    "Nome": ["Bruno", "Luiz", "Raquel", "Luan", "Ana", "Lorena", "Tiago", "Mylena", "Cyntia", "Pamela"],
    "Idade": ["40", "25", "45", "20", "70", "18", "55", "17", "22", "40"],
    "Amputação": ["Transfemoral", "Transradial", "Transfemoral", "Transfemoral", "Osteointegração", "Osteointegração", "Transtibial", "Transfemoral", "Transtibial", "Osteointegração"]
})
```

```python
amputee_df.head()
```

```python
#Chamando o modelo LLM da OpenAi

OPENAI_API_KEY = ""
llm = OpenAI(api_token=OPENAI_API_KEY)
```

```python
#Chamando o dataframe e enviando o comando de prompt

pandas_ai = PandasAI(llm)
pandas_ai.run(amputee_df, prompt='Mostre quem possuem as 3 maiores idades e classifique')
```

```python
#Prompt para plotar gráficos

pandas_ai.run(amputee_df, "Classifique por nome e idade e mostre a amputação de cada uma pessoa separadamente, criando um gráfico plot")
```
